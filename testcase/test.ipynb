```python
import json
import requests
import datetime as Date

INPUT_FILE = 'djia_qna (1).json'
OUTPUT_FILE = f'results_output_{Date.datetime.now().strftime("%Y%m%d_%H%M%S")}.jsonl'  # JSON Lines format (m·ªói d√≤ng = 1 JSON object)
API_URL = "http://localhost:8000/api/ask"

def extract_clean_answer(api_response: dict) -> str:
    answer = api_response.get("answer", "")
    lines = answer.strip().splitlines()
    lines = [line for line in lines if not line.strip().startswith("#")]
    return " ".join(line.strip() for line in lines if line.strip())

with open(INPUT_FILE, 'r', encoding='utf-8') as f_input, open(OUTPUT_FILE, 'a', encoding='utf-8') as f_output:
    questions_data = json.load(f_input)

    for item in questions_data[:]:
        question = item["question"]
        expected_answer = item["answer"]
        type = item["type"]
        expected_answer = item["answer"]
        complexity = item["complexity"]
        result = {
            "number": item["number"],
            "type": type,
            "complexity": complexity,
            "question": question,
            "expected_answer": expected_answer,
        }

        try:
            response = requests.post(API_URL, json={"text": question}, timeout=30)
            response.raise_for_status()
            data = response.json()
            result["generated_answer"] = extract_clean_answer(data)
        except Exception as e:
            result["generated_answer"] = f"[ERROR] {str(e)}"

        # Ghi t·ª´ng k·∫øt qu·∫£ v√†o file (m·ªói d√≤ng = 1 object JSON)
        f_output.write(json.dumps(result, ensure_ascii=False) + "\n")
        
        print(f"‚úÖ ƒê√£ x·ª≠ l√Ω c√¢u h·ªèi s·ªë {item['number']}")

print(f"\nüìÅ T·∫•t c·∫£ k·∫øt qu·∫£ ƒë√£ ƒë∆∞·ª£c l∆∞u v√†o {OUTPUT_FILE}")
```
